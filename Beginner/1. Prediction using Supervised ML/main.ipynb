```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv('student_scores.csv')
data.head()
```

```python
data.plot(x='Hours', y='Scores', style='o')
plt.title('Housr Vs Scores')
```

```python
sns.regplot(x=data['Hours'], y=data['Scores'])
plt.title('Relation Is Positive.')
```

```python
X = data.iloc[:, 0].values
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)

X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f'Model Score: {model.score(X_test, y_test):.3f} %',
      f'Model MSE  : {mean_squared_error(y_test, model.predict(X_test))}', 
      f'Model MAE  : {mean_absolute_error(y_test, model.predict(X_test))}', 
      
      
      sep='\n')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Iris.csv')
df = df.iloc[:, 1:]
df.head()
```

```python
X = df.iloc[:, :-1]
```

```python
res = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, max_iter=1000, random_state=0)
    model.fit(X)
    res.append(model.inertia_)

plt.plot(range(1,11), res)
```

```python
df['Species'].unique()
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
pred = model.predict(X)
```

```python
color_label = list(zip('g b y'.split(' '), df['Species'].unique()))

for i in range(3):
    plt.scatter( X.iloc[ pred == i, 0],
                 X.iloc[ pred == i, 1], 

                 c = color_label[i][0], 
                label = color_label[i][1] )
    
plt.scatter( model.cluster_centers_[:, 0],
             model.cluster_centers_[:, 1], 
             c='r', label="Centriods" )

plt.legend()
plt.title("3 Clusters and Its Centroids")
```
